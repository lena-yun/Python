## 왜 크롤링이 필요한지와 크롤링 준비 & 이미지 스크래핑 &  자연어처리

- 이미지 스크래핑

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.error   import HTTPError
from urllib.error   import URLError
from urllib.parse   import quote_plus
```

```python
base_url = 'https://search.naver.com/search.naver?where=image&sm=tab_jum&query='

#동적으로 키워드 입력받도록
keyword = input('검색어 입력 : ')
imgCnt  = int(input('스크랩 할 이미지 개수 입력 : '))


try :
    url  = base_url + quote_plus(keyword)
    ## quote_plus 문자어 처리...??
    html = urlopen(url)
except HTTPError as he :
    print('http error')
except URLError as ue :
    print('url error')
else :
    soup = BeautifulSoup(html.read(), 'html.parser', from_encoding='utf-8')
```

```python
img = soup.find_all('img', '_img')
```

```python
img
```

```python
# imgURL부분만 추출
cnt = 1
for i in img :
    imgUrl = i['data-source']
#    print(imgUrl)
    with urlopen(imgUrl) as file :
        with open('./images/' + str(cnt) +'.jpg', 'wb') as imgFile:
            img = file.read()
            imgFile.write(img)
    cnt += 1
    if cnt > imgCnt :
        break
print('image download sucess')
        # wb : write인데 binary형태로 진행
```

```python
from time import sleep
from random import randint
import pandas as pd
from IPython.core.display import clear_output
from time import time
from bs4 import BeautifulSoup
from requests import get
```

```python
## loop 사이에 request가 잘 진행되고 있는지를 보기 위한 indicatot sentence
start_time = time()
#start_time

requests = 0
for _ in range(5) :
    requests += 1
    sleep(randint(1,3))
    current_time = time()
    elapsed_time = current_time - start_time
    print("request : {}, frequency : {}".format(requests, requests/elapsed_time))
    
    clear_output(wait=True) # 데이터가 추가되어 보이지 않고 바뀌는 형태로 출력
```

```python
base_url = 'https://search.naver.com/search.naver?where=post&sm=tab_jum&query='
keyword = input('검색어 입력 : ')

try :
    url  = base_url + quote_plus(keyword)
    html = urlopen(url)
except HTTPError as he :
    print('http error')
except URLError as ue :
    print('url error')
else :
    soup = BeautifulSoup(html.read(), 'html.parser', from_encoding='utf-8')
```

```python
pages = [str(i) for i in range(11)]
pages
```

```python
# 내가 한 방법 ############

price_list = []
title_list = []
link_list  = []

## loop 사이에 request가 잘 진행되고 있는지를 보기 위한 indicatot sentence
start_time = time()
requests = 0

for page in pages :
    requests += 1
    url = get('https://search.shopping.naver.com/search/all?frm=NVSHATC&origQuery=%ED%94%84%EB%9D%BC%EB%8B%A4&pagingIndex='+page+'&pagingSize=40&productSet=total&query=%ED%94%84%EB%9D%BC%EB%8B%A4&sort=rel&timestamp=&viewType=list')
    sleep(randint(1,3))
    current_time = time()
    elapsed_time = current_time - start_time
#    print("request : {}, frequency : {}".format(requests, requests/elapsed_time))   
#    clear_output(wait=True) # 데이터가 추가되어 보이지 않고 바뀌는 형태로 출력
    soup = BeautifulSoup(url.text, 'html.parser')
    
    
    all = soup.find_all('a', {'class' : 'basicList_link__1MaTN'})
    for i in all:
        # title, link가져오기
        print(i.attrs['title'])
        title_list.append(i.attrs['title'])   
    
        print(i.attrs['href'])
        link_list.append(i.attrs['href'])   
        
    # price 가져오기
    prices = soup.find_all('span', 'price_num__2WUXn')
    for i in prices :
        print(i.getText())
        price_list.append(i.getText())
    
#print(title_list)
```

```python
print(len(title_list))
print(len(link_list))
print(len(price_list))
```

```python
#####강사님

price_list = []
title_list = []
link_list  = []

## loop 사이에 request가 잘 진행되고 있는지를 보기 위한 indicatot sentence
start_time = time()
requests = 0

for page in pages :
    requests += 1
    url = get('https://search.shopping.naver.com/search/all?frm=NVSHATC&origQuery=%ED%94%84%EB%9D%BC%EB%8B%A4&pagingIndex='+page+'&pagingSize=40&productSet=total&query=%ED%94%84%EB%9D%BC%EB%8B%A4&sort=rel&timestamp=&viewType=list')
    
    # request 에 delaytime 을 준다.
    sleep(randint(1,3))
    current_time = time()
    elapsed_time = current_time - start_time
#    print("request : {}, frequency : {}".format(requests, requests/elapsed_time))   
#    clear_output(wait=True) # 데이터가 추가되어 보이지 않고 바뀌는 형태로 출력
    soup = BeautifulSoup(url.text, 'html.parser')
    
    
    # title, link, price
    # 가격 가져오기
    prices = soup.select('.price_num__2WUXn')
    prices_text = [price.text for price in prices]
    price_list.extend(prices_text)
    print(prices_text)
    
    # 타이틀, 링크 가져오기
    titles = soup.select('.basicList_link__1MaTN')
    titles_text = [title.text for title in titles]
    title_list.extend(titles_text)
    print(titles_text)
    
    link_text = [title['href'] for title in titles]
    link_list.extend(link_text)
    print(link_text)

#print(title_list)
```

```python
print(len(title_list))
print(len(link_list))
print(len(price_list))
```

```python
prada_page_df = pd.DataFrame({
    'title' : title_list,
    'price' : price_list,
    'link' : link_list,
    
})
prada_page_df
```

```python
prada_page_df.to_csv('./data/prada_page_df.csv', mode='w', encoding='utf-8', index=False)
print('success')
```
